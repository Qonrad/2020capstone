```python
import pandas as pd
from IPython.core.display import display, HTML

all_bids_ids = pd.read_csv("nfb_subset_SUBJECT_IDS.csv")
all_bids_ids['ID'] = all_bids_ids['ID'].str[4:]
all_bids_ids['conn_ID'] = all_bids_ids.index + 1

all_score_ids = pd.read_csv("events_IDs.csv")

bids_subset_ids = pd.merge(all_score_ids, all_bids_ids, how='inner', on='ID').sort_values(by=['ID']).reset_index(drop=True)
display(bids_subset_ids)
```

```python
from scipy.io import loadmat
import os
import pandas as pd
import numpy as np

path = './new_matrices'
files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if 'resultsROI_Subject' in file:
            files.append(os.path.join(r, file))
#print(files)


X = np.array([np.zeros(188805)]) #quite possibly unnecessarily complicated, unsure how to do this
#it absolutely must be checked that the scores match up properly with the correct connectivity matrix

for i in range(len(files)):
    if int(files[i][33:36]) in bids_subset_ids.conn_ID.values:
        connectome = loadmat(files[i])
        mat = connectome['Z']
        mat = mat[:,0:len(mat[0]) - 3]
        brief = np.array([])
        for j in range(len(mat)):
            #print(mat[j][j+1:])
            brief = np.concatenate((brief, mat[j][j+1:]), axis=0)
        temp = np.array([brief.tolist()])
        X = np.concatenate((X, temp), axis=0)
    else:
        print(int(files[i][33:36]), "has no score. Not reading their matrix.")
X = X[1:]
print(len(X))
np.save("connectomes", X)
```
